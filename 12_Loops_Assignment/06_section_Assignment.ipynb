```python
# Section 6: Combined Challenges
# Problem 6.1: Prime Numbers in Range (Medium)
# Write a program to print all prime numbers between two given numbers (inclusive).

# Example: Between 10 and 30: 11, 13, 17, 19, 23, 29

# Hint: Outer loop for each number in range. For each number, use inner loop to check if it's prime (like Problem 2.5).
```

```python
for i in range(10,30):
    for j in range(2,i):
        if i%j==0:
            print(f"{i} Not a prime Number")
            break
    else:
        print(f"{i} Prime number")
```

```python
# Problem 6.2: GCD Calculator (Medium)
# Write a program to find the Greatest Common Divisor (GCD) of two numbers using the Euclidean algorithm.

# Algorithm:

# 1 Divide the larger number by the smaller number
# 2 Replace the larger number with the smaller number
# 3 Replace the smaller number with the remainder
# 4 Repeat until the remainder is 0
# 5 The GCD is the last non-zero remainder
# Hint: While b != 0: store remainder (a%b), then a=b, b=remainder. When loop ends, a is the GCD.
```

```python
num_1 = int(input("Enter Your Number:-"))

num_2 = int(input("Enter Your Number:-"))

while num_2 > 0:
    remainder = num_1 % num_2
    
    num_1 = num_2

    num_2 = remainder

print("GCD is", num_1)

    
```

```python
# Problem 6.3: Perfect Number Checker (Hard)
# Write a program to check if a number is a perfect number.

# A perfect number is a positive integer that is equal to the sum of its proper positive divisors (excluding the number itself).

# Example: 6 is perfect because 6 = 1 + 2 + 3

# Example: 28 is perfect because 28 = 1 + 2 + 4 + 7 + 14

# Hint: Loop from 1 to n-1. If i divides n (n%i==0), add i to sum. Compare sum with n.
```

```python
num= int(input("Enter your Number:-"))

count=0

for i in range(1,num):
    if num%i == 0:
        count+=i

if count == num:
    print(f"{num} is a perfect number")    
else:
    print(f"{num} is not a perfect number")
```

```python
# Problem 6.4: LCM Calculator (Medium)
# Write a program to find the Least Common Multiple (LCM) of two numbers.

# Hint: LCM(a, b) = (a × b) / GCD(a, b)

# You can use the GCD function from Problem 6.2.

# Hint: First calculate GCD using Euclidean algorithm (Problem 6.2), then use formula: LCM = (a * b) // GCD
```

```python

# Problem 6.5: Number to Words (Hard)
# Write a program that converts a number (0-999) to words.

# Examples:

# 0: Zero
# 25: Twenty Five
# 123: One Hundred Twenty Three
# 500: Five Hundred
# Hint: Create lists for ones, teens, tens, etc. Extract hundreds, tens, ones digits. Convert each part to words and combine.
```

```python
# Lists for words
ones = ["", "One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine"]
teens = ["Ten", "Eleven", "Twelve", "Thirteen", "Fourteen", "Fifteen",
         "Sixteen", "Seventeen", "Eighteen", "Nineteen"]
tens = ["", "", "Twenty", "Thirty", "Forty", "Fifty", "Sixty", "Seventy", "Eighty", "Ninety"]

num = int(input("Enter a number (0-999): "))

words = ""

if num == 0:
    words = "Zero"
else:
    # Hundreds
    if num >= 100:
        words += ones[num // 100] + " Hundred"
        num = num % 100
        if num > 0:
            words += " "

    # Tens and Ones
    if num >= 20:
        words += tens[num // 10]
        if num % 10 > 0:
            words += " " + ones[num % 10]
    elif num >= 10:
        words += teens[num - 10]
    elif num > 0:
        words += ones[num]

print("In words:", words)

```
